```python
# Plot Arrests by Perpetrator Race
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='PERP_RACE', order=df['PERP_RACE'].value_counts().index, palette='magma')
plt.title("Arrests by Perpetrator Race")
plt.xlabel("Count")
plt.ylabel("Race")
plt.show()
```

Observation : Black Race is the highest number of perpetrator


```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, x='AGE_GROUP', order=df['AGE_GROUP'].value_counts().index,palette='Set2')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=2, color='black', fontsize=10)

plt.title("Arrests by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

Observation: Highest number of perpetartors are between 25-44 age group.

```python
df['ARREST_DATE'] = pd.to_datetime(df['ARREST_DATE'], errors='coerce')

# Extract month names from the date
df['ARREST_MONTH'] = df['ARREST_DATE'].dt.month_name()

# Count arrests per month
month_counts = df['ARREST_MONTH'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(month_counts, labels=month_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"), startangle=90)
plt.title("Arrests Distribution by Month")
plt.show()
```

Obsevation : The distribution is pretty equal suggesting every month there are close to equal number of arrests.

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='PERP_RACE', hue='PERP_SEX', palette='viridis')

# Add labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, color='black', fontsize=10)

# Titles and labels
plt.title("Arrests by Perpetrator Race and Sex")
plt.xlabel("Perpetrator Race")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="Perpetrator Sex")

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='LAW_CAT_CD', hue='PERP_SEX', palette='Set2')

# Add labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3, color='black', fontsize=9)

# Titles and labels
plt.title("Arrests by Law Category and Perpetrator Sex")
plt.xlabel("Law Category")
plt.ylabel("Count")
plt.legend(title="Perpetrator Sex")
plt.show()
```

```python
# Bar chart for ARREST_BORO distribution
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='ARREST_BORO' ,hue='PERP_SEX',palette='husl')

# Add labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=1, color='black', fontsize=10)

# Titles and labels
plt.title("Arrests by Borough")
plt.xlabel("Borough")
plt.ylabel("Count")
plt.show()
```

Observation : K borough has the highest number of arrests.

```python
# Group by date and borough
arrest_trends = df.groupby(['ARREST_DATE', 'ARREST_BORO']).size().unstack()

# Plot the trends
plt.figure(figsize=(12, 6))
arrest_trends.plot(ax=plt.gca())
plt.title("Daily Arrest Trends by Borough")
plt.xlabel("Date")
plt.ylabel("Number of Arrests")
plt.legend(title="Borough")
plt.grid()
plt.show()
```

```python
# Top 10 most common offenses
top_offenses = df['OFNS_DESC'].value_counts().index[:10]
df_top = df[df['OFNS_DESC'].isin(top_offenses)]

# Grouped bar chart
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df_top, y='OFNS_DESC', hue='PERP_SEX', palette='Set2')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3, color='black', fontsize=8, weight='bold')

plt.title("Top 10 Offense Types by Perpetrator Sex")
plt.xlabel("Count")
plt.ylabel("Offense Description")
plt.legend(title="Perpetrator Sex")
plt.show()
```

```python
#Arrest patterns by jurisdiction code
jurisdiction_distribution = df.groupby(['JURISDICTION_CODE', 'OFNS_DESC']).size().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(jurisdiction_distribution.fillna(0), cmap='coolwarm', linewidths=0.5)
plt.title("Arrest Patterns by Jurisdiction Code")
plt.xlabel("Offense Type")
plt.ylabel("Jurisdiction Code")
plt.show()
```

```python
# Severity of charges by racial group
charge_severity = df.groupby(['PERP_RACE', 'LAW_CAT_CD']).size().unstack()
charge_severity.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title("Charge Severity (Felony vs. Misdemeanor) by Race")
plt.xlabel("Race")
plt.ylabel("Number of Arrests")
plt.legend(title="Charge Severity")
plt.grid()
plt.show()
```
